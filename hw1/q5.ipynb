```python
#Yael
def computeTF(token: str, doc: str):
  count = 0
  doc = doc.split()
  if len(doc) == 0:
    return 0

  for word in doc:
      if word == token:
          count += 1
  return count/ len(doc)
print(computeTF("hello","hello my name is hello"))
```

```python
def computeDF(token: str, corpus: list):
  counter = 0
  for doc in corpus:
    if computeTF(token, doc) > 0:
      counter += 1
  return counter

print(computeDF("hello",["hello my name is hello","Hey dude","hello you","check"]))
```

```python
def computeIDF(token: str, corpus: list):
  import numpy as np
  return np.log(len(corpus)/(computeDF(token,corpus) + 1))

print(computeIDF("hello",["hello my name is hello","Hey dude","hello you","check"]))
```

```python
def computeTFIDF(token: str, doc: str, corpus: str):
  return computeTF(token, doc) * computeIDF(token, corpus)
```

```python
# constructing the corpus

from datasets import load_dataset

# loading wikipedia using HuggingFace
wikipedia = load_dataset('wikipedia', '20220301.en', beam_runner='DirectRunner')

# Searching for the New Zealand's wikipedia passage
for passage in wikipedia['train']:
  if passage['title'] == 'New Zealand':
    new_zealand_passage = passage['text']
    break

# Breaking the passage into a corpus of sentences.
corpus = new_zealand_passage.split('.')
print(corpus)
```

In order to retrieve the most relevant sequences out of the corpus, to the query: What is the capital of New Zealand?. As a final answer, please print the 3 highest scored sentences out of the corpus.


```python
# defining the query
from statistics import mean
grades = [0 for i in range(len(corpus))]
query = 'What is the capital of New Zealand?'
tokens = query.split()
for index, sentense in enumerate(corpus):
  grades[index] = mean([computeTFIDF(token, sentense, corpus) for token in tokens])
grades = [(i,x) for i,x in enumerate((grades))]
grades.sort(reverse=True,  key=lambda x: x[1])
#print(grades)
best = grades[:3]
#print(best)
for b in best:
  print(corpus[b[0]])
  print("\n")
```
